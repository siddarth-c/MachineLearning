# <u>***Complaint Detection***

This notebook is divided into 3 parts. <br>
1. Simple complaint detection using XLNet
2. Simple sentiment analysis of amazon reviews
3. Combining both models

```
from simpletransformers.classification import ClassificationModel
import logging

logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)

import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
```

<u>**Part 1 : Complaint Detection using XLNet**

```
df1 = pd.read_csv('cd.csv')
df1.head(2)
```

```
X1 = df1['tweet']
y1 = df1['label']

X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=4, stratify=y1)

train_df1 = pd.DataFrame(X_train1)
train_df1['target'] = y_train1

eval_df1 = pd.DataFrame(X_test1)
eval_df1['target'] = y_test1
```

```
model1 = ClassificationModel('xlnet', 'xlnet-base-cased', args={'num_train_epochs':2, 'train_batch_size':32, 'max_seq_length':128})

# Train the model
model1.train_model(train_df1)

# Evaluate the model
result1, model_outputs1, wrong_predictions1 = model1.eval_model(eval_df1, acc=sklearn.metrics.accuracy_score)
```

<u>**Part 2 : Sentiment Analysis using XLNet**

```
df2 = pd.read_csv('sentiment.csv', encoding = 'unicode_escape')
df2.head(2)
```

```
df2 = df2.drop('Unnamed: 0', axis = 1)

X2 = df2['reviews']
y2 = df2['ratings']

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=42, stratify=y2)

train_df2 = pd.DataFrame(X_train2)
train_df2['target'] = y_train2

eval_df2 = pd.DataFrame(X_test2)
eval_df2['target'] = y_test2
```

```
model2 = ClassificationModel('xlnet', 'xlnet-base-cased', args={'num_train_epochs':2, 'train_batch_size':32, 'max_seq_length':128, 'overwrite_output_dir': True})

# Train the model
model2.train_model(train_df2)

# Evaluate the model
result2, model_outputs2, wrong_prediction2s = model2.eval_model(eval_df2, acc=sklearn.metrics.accuracy_score)
```

<u>**Part 3 : Combining both models**

```
predictions, raw_outputs = model2.predict(train_df1['tweet'].tolist())
```

```
for i in range(len(predictions)):
  if predictions[i] == 1:
    df1.at[i, 'tweet'] = 'Good' + df1.at[i, 'tweet']
  else:
    df1.at[i, 'tweet'] = 'Bad' + df1.at[i, 'tweet']
```

```
X3 = df1['tweet']
y3 = df1['label']

X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.2, random_state=4, stratify=y3)

train_df3 = pd.DataFrame(X_train3)
train_df3['target'] = y_train3

eval_df3 = pd.DataFrame(X_test3)
eval_df3['target'] = y_test3
```

```
model3 = ClassificationModel('xlnet', 'xlnet-base-cased', args={'num_train_epochs':2, 'train_batch_size':32, 'max_seq_length':128, 'overwrite_output_dir': True})


# Train the model
model3.train_model(train_df3)

# Evaluate the model
result3, model_outputs3, wrong_predictions3 = model3.eval_model(eval_df3, acc=sklearn.metrics.accuracy_score)
```

```
print('Result from Part 1 :', result1)
print('Result from Part 3 :', result3)
```

```
ones = sum(eval_df3['target'] == 1)
zeros =  len(eval_df3) - ones
print('Number of +ve classes in evaluation set :', ones)
print('Number of -ve classes in evaluation set :', zeros)
```

Due to the class imbalance, the area under the precision-recall curve (auprc) would be a better measure than Accuracy (auc). Note that by adding just 'Good' or 'Bad', the model's auprc improved by a small margin.
