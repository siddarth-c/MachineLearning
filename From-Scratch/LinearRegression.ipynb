#### Least Square Method

```python

def PolyCoefficients(x, coeffs):

    o = len(coeffs)
    y = 0
    for i in range(o):
        y += coeffs[i]*x**i
    return y

def matrix_g(deg, x):    
    ls = []
    for j in range(int(deg)+1):
        ls.append(pow(x,j))
    return np.stack(ls).T


d = y
G = matrix_g(deg, x)

p1 = np.linalg.inv(np.dot(G.T,G))
M = np.dot(np.dot(p1, G.T), d)

x_d = np.linspace(0, 11, 40)
plt.plot(x_d, PolyCoefficients(x_d, M))
plt.scatter(x, y)
plt.show()
```
