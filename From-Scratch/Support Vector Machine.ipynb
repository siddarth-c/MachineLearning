```python
import numpy as np

# For plotting in dark mode
from jupyterthemes import jtplot
import matplotlib.pyplot as plt
jtplot.style()
```

```python
n = 20
cl1 = np.random.random((1,n))
cl2 = np.random.random((1,n)) + 1.5
```

```python
plt.scatter(cl1, range(n))
plt.scatter(cl2, range(n))
plt.show()
```

```python
def data_generate(size, feat):
    
    x = np.random.random((size, feat))
    x[ : - size//2] = x[:-size//2] + 0.6
    y0 = np.zeros((1, size//2))
    y1 = np.ones((1, size//2))
    y = np.hstack((y0, y1))
    return x, y
```

```python
trainx, trainy = data_generate(30, 5)
```

```python
plt.scatter(trainx.T[2][:15], trainx.T[3][:15])
plt.scatter(trainx.T[2][15:], trainx.T[3][15:])
plt.show()
```
