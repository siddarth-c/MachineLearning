# OpenCV and matplotlib

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

from PIL import Image
import glob
import tensorflow as tf
from tensorflow import keras

img = cv2.imread('train/1/013.jpg') # Load image
```

### https://stackoverflow.com/questions/15072736/extracting-a-region-from-an-image-using-slicing-in-python-opencv/15074748#15074748 <br>
OpenCV follows BGR order while matplotlib follows RGB <br>
So to print the image using matplotlib (which will be used frequently we have invert the layers)


#### Method 1

```python
b,g,r = cv2.split(img) # splits the image into 3 channels
img2 = cv2.merge([r,g,b]) 
plt.subplot(121);plt.imshow(img) # expects distorted color
plt.subplot(122);plt.imshow(img2) # expect true color
plt.show
```

#### Method 2

```python
img3=img[..., ::-1] # reverses the order of the channels
plt.imshow(img3)
```

### Accessing and Modifying pixel values

```python
img2=img3.copy()
img2[100]=[255,255,255]
plt.imshow(img2)
```

Note that the whole line 100's pixel value is set to 0

### Selecting part of an image

```python
img2=img3.copy()
beak = img2[50:80,15:40]
plt.imshow(beak)
```

### Reading an entire folder of images

```python
image = []
input_arr=[]

for filename in glob.glob('train/1/*.jpg'): # read every .jpg inside train/1/
    im=Image.open(filename) # read image
    image.append(im) # append image to image list
    input_arr.append(tf.keras.preprocessing.image.img_to_array(im)) # convert image into array and append to list
```

```python
img4=tf.keras.preprocessing.image.array_to_img(input_arr[0])
img4
```
